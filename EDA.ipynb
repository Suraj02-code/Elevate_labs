```python
sns.histplot(df['Age'], kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
sns.boxplot(x=df['Fare'])
plt.title('Boxplot of Fare')
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Count of Survival (0 = No, 1 = Yes)')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival Count Based on Gender')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title('Survival Count Based on Passenger Class')
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Passenger Class vs Age')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']])
plt.show()
```
